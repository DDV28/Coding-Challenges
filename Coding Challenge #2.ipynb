# Coding Challenge Instructions:

function that takes rand number as input
call this function 100 times to ensure its working (give different rand number between 1-100 as input each time)
if rand number < 3, display "weird"
if number between 4 and 7 display "not weird"
if number greater than 7 and odd display "weird"
otherwise display "not weird"

```python
import random
```

```python
def rand(num):
    print("number:", num)
    
    if num < 3:
        print("weird")
    elif num <= 7:
        print("not weird")
    elif num % 2 != 0:
        print("weird")
    else:
        print("not weird")
```

```python
for i in range(100):
    num = random.randint(1, 100)
    rand(num)
```
